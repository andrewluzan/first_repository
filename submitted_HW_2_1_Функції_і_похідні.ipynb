```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Let us declare symbolic value 'x', function and find its derivative
x = sp.symbols('x')
f = 10 * x - 5
derivative_f = sp.diff(f, x)
print("Derivative of function:", derivative_f)
numerical_func = sp.lambdify(x, derivative_f, "numpy")
```

```python
# Let us build graphs of the function and its derivative

x = np.linspace(0, 10, 10)
f = 10 * x - 5
#derivative_f = numerical_func(x)
derivative_f = np.full_like(x, numerical_func(0)) # Make derivative match x_vals' shape

# Creation of graph
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x-5', color='blue')
plt.plot(x, derivative_f, label="f'(x) = 10", color='green')

#  Making dots and corresponding tangents
for x_point in x:
  f_point = 10*x_point -5
  df_point = 10
  tangent_line = df_point * (x - x_point) + f_point  # Equation of tangent: y - y1 = m(x - x1)

  plt.scatter(x_point, f_point, color='red') # dot at curve
  plt.plot(x, tangent_line, '--', color='red', label='Tangent') # tangent

# Adding title and legend
plt.title("Graph of a function and its derivative")
plt.xlabel("x")
plt.ylabel("f_lin")
plt.legend()
plt.grid(True)

plt.show()
```

### 2.
Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Define the symbolic variable and function
x = sp.symbols('x')
f = sp.sin(x)
derivative_f = sp.diff(f, x)

print("Derivative of f(𝑥):", derivative_f)

# Convert symbolic expressions to numerical functions
f_numeric = sp.lambdify(x, f, 'numpy')
derivative_f_numeric = sp.lambdify(x, derivative_f, 'numpy')

# Define the range for x values
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 500)
f_values = f_numeric(x_values)
derivative_f_values = derivative_f_numeric(x_values)

# Plot the functions
plt.figure(figsize=(10, 6))

# Plot the original function
plt.plot(x_values, f_values, label='f(x) = sin(x)', linewidth=2)

# Plot the derivative
plt.plot(x_values, derivative_f_values, label="f'(x) = cos(x)", linewidth=2, linestyle='--')

# Add labels, legend, and grid
plt.title('Function sin(x) and Its Derivative', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```

```python
# Define the symbolic variable and function
x = sp.symbols('x')
f = sp.ln(x)
derivative_f = sp.diff(f, x)

print("Derivative of f(𝑥):", derivative_f)

# Convert symbolic expressions to numerical functions
f_numeric = sp.lambdify(x, f, 'numpy')
derivative_f_numeric = sp.lambdify(x, derivative_f, 'numpy')

# Define the range for x values
x_values = np.linspace(0.01, 10, 50)
f_values = f_numeric(x_values)
derivative_f_values = derivative_f_numeric(x_values)

# Plot the functions
plt.figure(figsize=(10, 6))

# Plot the original function
plt.plot(x_values, f_values, label='f(x) = ln(x)', linewidth=2)

# Plot the derivative
plt.plot(x_values, derivative_f_values, label="f'(x) = 1/x", linewidth=2, linestyle='--')

# Add labels, legend, and grid
plt.title('Function ln(x) and Its Derivative', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.grid(alpha=0.4)
plt.legend(fontsize=12)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Define the symbolic variable and function
x = sp.symbols('x')
f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)

print("Derivative of f(𝑥):", derivative_f)

# Convert symbolic expressions to numerical functions
f_numeric = sp.lambdify(x, f, 'numpy')
derivative_f_numeric = sp.lambdify(x, derivative_f, 'numpy')

# Define the range for x values
x_values = np.linspace(0, 100, 10)
f_values = f_numeric(x_values)
derivative_f_values = derivative_f_numeric(x_values)


# Plot the original function
plt.figure(figsize=(10, 5))
plt.plot(x_values, f_values, label='f(x) = x^2-5x+1', color='blue', linewidth=2)


# Making dots and corresponding tangents
for x_point in x_values:
  f_point = x_point**2 - 5*x_point -1
  df_point = 2*x_point + 5
  tangent_line = df_point * (x_values - x_point) + f_point  # Equation of tangent: y - y1 = m(x - x1)

  plt.scatter(x_point, f_point, color='red') # dot at curve
  plt.plot(x_values, tangent_line, '--', color='red', label='Tangent') # tangent

# Adding a legend, labels, and title
plt.title("Graph of a function and its v derivative")
plt.xlabel("x")
plt.ylabel("f_lin")
plt.legend()
plt.grid(True)

plt.show()
```
